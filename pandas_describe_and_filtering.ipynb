```python
#filtering rows on single condfition
high_salary =df[df["Salary"] > 50000]
print('Employees with salary greater than 50000:')
print(high_salary)
#
```

```python
#now mutiple condition on rows
#filtering rows salary > 50k & age > 30
filter = df[(df['Salary'] > 50000  ) & (df['Age']> 30)]
print ("Employee list Age > 30 + Salary  >50000")
print(filter)
```

```python
#using or condition (one condition can be True)
filter_or = df[(df['Age'] > 35)| (df['Performance score'] > 90)]
print("Employee list Age > 35 or Performance score > 90")
print(filter_or)
```
